```python
#encoding the data
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output= False,drop='first')
encoded= ohe.fit_transform(df[['Gender',
                               'Parental_Involvement',
                               'Access_to_Resources',
                               'Teacher_Quality',
                               'Learning_Disabilities']])
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(['Gender',
                               'Parental_Involvement',
                               'Access_to_Resources',
                               'Teacher_Quality',
                               'Learning_Disabilities']))
encoded_df.head()
df=df.drop(['Gender',
            'Parental_Involvement',
            'Access_to_Resources',
            'Teacher_Quality',
            'Learning_Disabilities'],axis=1)
df=pd.concat([df,encoded_df],axis=1)
df.head()
```

```python
x = df.drop("Exam_Score", axis=1)
y = df["Exam_Score"]
```

```python
#model selection for the data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# The one-hot encoding has already been applied in a previous cell (nRUUxy_f8bgm),
# and the original categorical columns have been dropped from 'df'.
# Re-applying it here causes a KeyError because the columns no longer exist.
# We will use the 'df' that has already been transformed.

# Define x and y from the correctly prepared df, selecting only the features
# that were used in the linear regression model.
x = df[['Hours_Studied',
        'Attendance',
        'Sleep_Hours',
        'Previous_Scores',
        'Tutoring_Sessions',
        'Gender_Male',
        'Parental_Involvement_Low',
        'Parental_Involvement_Medium',
        'Access_to_Resources_Low',
        'Access_to_Resources_Medium',
        'Teacher_Quality_Low',
        'Teacher_Quality_Medium',
        'Learning_Disabilities_Yes'
]]
y = df["Exam_Score"]

x_train,x_test,y_tarin,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler= StandardScaler()
x_train=scaler.fit_transform(x_train) # Fit and transform on x_train
x_test=scaler.transform(x_test)       # Only transform x_test using the scaler fitted on x_train
```

```python
# giving x and y values
x = df[['Gender_Male' ,'Parental_Involvement_Low' ,'Parental_Involvement_Medium',
 'Access_to_Resources_Low' ,'Access_to_Resources_Medium',
 'Teacher_Quality_Low' ,'Teacher_Quality_Medium',
 'Learning_Disabilities_Yes','Hours_Studied',
    'Attendance',
    'Sleep_Hours',
    'Previous_Scores',
    'Tutoring_Sessions'
]]

y = df['Exam_Score']
```

```python

numeric_columns = [
    'Hours_Studied',
    'Attendance',
    'Sleep_Hours',
    'Previous_Scores',
    'Tutoring_Sessions'
]

# The error occurs because x_train and x_test are numpy arrays,
# and numpy arrays cannot be indexed by column names (strings).
# We need to get the integer indices corresponding to these column names
# from the original 'x' DataFrame (which is available in the kernel).

# Get the integer positions of the numeric columns from the 'x' DataFrame
col_indices = [x.columns.get_loc(col) for col in numeric_columns]

# Apply the scaler to these specific columns using integer indexing
x_train[:, col_indices] = scaler.fit_transform(x_train[:, col_indices])
x_test[:, col_indices] = scaler.transform(x_test[:, col_indices])
```

```python
print(type(x_train))
print(type(x_test))
print(x_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_tarin)
y_pred=model.predict(x_test)
```

```python
print(y_pred[:10])
print(y_test[:10])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("mae: ",mae)
print("mse:", mse)
print(r2)
```

```python
print("Intercept:", model.intercept_)
print("coefficient:", model.coef_)
print(df.columns)
feature_names = ['Gender_Male' ,'Parental_Involvement_Low' ,'Parental_Involvement_Medium',
                'Access_to_Resources_Low' ,'Access_to_Resources_Medium',
                'Teacher_Quality_Low' ,'Teacher_Quality_Medium',
                'Learning_Disabilities_Yes','Hours_Studied',
                'Attendance',
                'Sleep_Hours',
                'Previous_Scores',
                'Tutoring_Sessions']
coefficients = pd.DataFrame({
    "Feature": feature_names,
    "Coefficient": model.coef_
})

coefficients = coefficients.sort_values(
    by="Coefficient",
    ascending=False
)

print(coefficients)
```

```python
plt.figure(figsize=(10,6))
coefficients = coefficients.sort_values("Coefficient")
plt.barh(
    coefficients["Feature"],
    coefficients["Coefficient"]
)
plt.title("Feature Importance")
plt.show()
```

```python

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor

dt=DecisionTreeRegressor(random_state=42)
rf= RandomForestRegressor(n_estimators=100, random_state= 42)

gb=GradientBoostingRegressor(random_state=42)
```

```python

dt.fit(x_train, y_tarin)
rf.fit(x_train, y_tarin)
gb.fit(x_train, y_tarin)
```

```python
from sklearn.metrics import (mean_absolute_error,mean_squared_error,r2_score)
import numpy as np
```

```python
def evaluate_model(name, y_test, prediction):
    mae= mean_absolute_error(y_test, prediction)
    mse=mean_squared_error(y_test, prediction)
    rmse=np.sqrt(mse)
    
    r2= r2_score(y_test, prediction)
    
    print(f"\n{name}")
    print("-"*30)
    
    print("MAE :",round(mae,3))
    print("RMSE :", round(rmse,3))
    print("R2 :", round(r2,3))
    
    return[name, mae, rmse, r2]
```

```python
results = []
results.append(evaluate_model("Linear Regression",y_test,y_pred))
dt_pred= dt.predict(x_test)
results.append(evaluate_model("Decision Tree", y_test, dt_pred))
rf_pred= rf.predict(x_test)
results.append(evaluate_model("Random Forest", y_test, rf_pred))
gb_pred= gb. predict(x_test)
results.append(evaluate_model("Gradient Boosting", y_test, gb_pred))
```

```python
results_df=pd.DataFrame(results,
                        columns=["Model",
                                 "MAE",
                                 "RMSE",
                                 "R2 SCORE"])
results_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(
    results_df["Model"],
    results_df["R2 SCORE"]
)

plt.title("Model comparison (R2 Score)")
plt.xlabel("MODELS")
plt.ylabel("R2 SCORE")

plt.show()
```

```python
best_model= results_df.sort_values(by="R2 SCORE",
                                   ascending=False)
best_model
```
